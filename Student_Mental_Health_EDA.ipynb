```python
import pandas as pd

df = pd.read_csv("cleaned_student_mental_health (1).csv")

df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df["Choose your gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.show()
```

```python
df["Do you have Depression?"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Depression Analysis")
plt.ylabel("")
plt.show()
```

```python
df["Do you have Anxiety?"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Anxiety Analysis")
plt.ylabel("")
plt.show()
```

```python
df["Do you have Panic attack?"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Panic Attack Analysis")
plt.ylabel("")
plt.show()
```

```python
df["Age"].hist(bins=10)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns

df2 = df.copy()

le = LabelEncoder()

for col in df2.columns:
    df2[col] = le.fit_transform(df2[col].astype(str))

plt.figure(figsize=(10,8))

sns.heatmap(df2.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
print("EDA Findings")
print("1. Female students are more represented.")
print("2. Some students reported depression and anxiety.")
print("3. Age distribution is concentrated in a small range.")
print("4. Mental health factors show relationships with each other.")
print("5. Correlation heatmap helps identify influencing factors.")
```
